```python
labels = DataFrame.target_class.values

DataFrame.drop(["target_class"],axis=1,inplace=True)

features = DataFrame.values
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler(feature_range=(0,1))

features_scaled = scaler.fit_transform(features)
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(features_scaled,labels,test_size=0.2)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
from sklearn.linear_model import LogisticRegression
lr_model = LogisticRegression(random_state=42,solver="liblinear",C=1.6,penalty="l1")

lr_model.fit(x_train,y_train)

y_head_lr = lr_model.predict(x_test)

lr_score = lr_model.score(x_test,y_test)

print("Logistic Regression")
print('Confusion Matrix:')
print(confusion_matrix(y_test,y_head_lr))
print('Classification Report:')
print(classification_report(y_test,y_head_lr))
print('Accuracy:', accuracy_score(y_test,y_head_lr))
```

```python
from sklearn.tree import DecisionTreeClassifier
dc_model = DecisionTreeClassifier(random_state=42)

dc_model.fit(x_train,y_train)

y_head_dc = dc_model.predict(x_test)

dc_score = dc_model.score(x_test,y_test)

print("DecisionTreeClassifier")
print('Confusion Matrix:')
print(confusion_matrix(y_test,y_head_dc))
print('Classification Report:')
print(classification_report(y_test,y_head_dc))
print('Accuracy:', accuracy_score(y_test,y_head_dc))
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc_model = RandomForestClassifier(n_estimators=37,random_state=42,max_leaf_nodes=200,criterion="entropy")

rfc_model.fit(x_train,y_train)

y_head_rfc = rfc_model.predict(x_test)

rfc_score = rfc_model.score(x_test,y_test)

print("RandomForestClassifier")
print('Confusion Matrix:')
print(confusion_matrix(y_test,y_head_rfc))
print('Classification Report:')
print(classification_report(y_test,y_head_rfc))
print('Accuracy:', accuracy_score(y_test,y_head_rfc))
```

```python
from sklearn.naive_bayes import GaussianNB
nb_model = GaussianNB()

nb_model.fit(x_train,y_train)

y_head_nb = nb_model.predict(x_test)

nb_score = nb_model.score(x_test,y_test)

print("Naive Bayes")
print('Confusion Matrix:')
print(confusion_matrix(y_test,y_head_nb))
print('Classification Report:')
print(classification_report(y_test,y_head_nb))
print('Accuracy:', accuracy_score(y_test,y_head_nb))
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn_model = KNeighborsClassifier(n_neighbors=7,weights="distance")

knn_model.fit(x_train,y_train)

y_head_knn = knn_model.predict(x_test)

knn_score = knn_model.score(x_test,y_test)

print("KNN")
print('Confusion Matrix:')
print(confusion_matrix(y_test,y_head_knn))
print('Classification Report:')
print(classification_report(y_test,y_head_knn))
print('Accuracy:', accuracy_score(y_test,y_head_knn))
```

```python
from sklearn.svm import SVC
svm_model = SVC(random_state=42,C=250,gamma=1.6,kernel="poly",probability=True)

svm_model.fit(x_train,y_train)

y_head_svm = svm_model.predict(x_test)

svm_score = svm_model.score(x_test,y_test)

print("Svm com Kernel Polynomial")
print('Confusion Matrix:')
print(confusion_matrix(y_test,y_head_svm))
print('Classification Report:')
print(classification_report(y_test,y_head_svm))
print('Accuracy:', accuracy_score(y_test,y_head_svm))
```

```python
svm_model_linear = SVC(kernel="linear")

svm_model_linear.fit(x_train,y_train)

y_head_svm_linear = svm_model_linear.predict(x_test)

svm_score_linear = svm_model.score(x_test,y_test)

print("Svm sem kerel")
print('Confusion Matrix:')
print(confusion_matrix(y_test,y_head_svm_linear))
print('Classification Report:')
print(classification_report(y_test,y_head_svm_linear))
print('Accuracy:', accuracy_score(y_test,y_head_svm_linear))
```

```python
from sklearn.metrics import confusion_matrix

cm_lr = confusion_matrix(y_test,y_head_lr)
cm_dc = confusion_matrix(y_test,y_head_dc)
cm_knn = confusion_matrix(y_test,y_head_knn)
cm_nb = confusion_matrix(y_test,y_head_nb)
cm_rfc = confusion_matrix(y_test,y_head_rfc)
cm_svm = confusion_matrix(y_test,y_head_svm)
cm_svm_linear = confusion_matrix(y_test, y_head_svm_linear)
```

```python
plt.figure(figsize=(24,12))

plt.suptitle("Confusion Matrixes",fontsize=24)

plt.subplot(2,3,1)
plt.title("Logistic Regression Confusion Matrix")
sns.heatmap(cm_lr,cbar=False,annot=True,cmap="CMRmap_r",fmt="d")

plt.subplot(2,3,2)
plt.title("Decision Tree Classifier Confusion Matrix")
sns.heatmap(cm_dc,cbar=False,annot=True,cmap="CMRmap_r",fmt="d")

plt.subplot(2,3,3)
plt.title("K Nearest Neighbors Confusion Matrix")
sns.heatmap(cm_knn,cbar=False,annot=True,cmap="CMRmap_r",fmt="d")

plt.subplot(2,3,4)
plt.title("Naive Bayes Confusion Matrix")
sns.heatmap(cm_nb,cbar=False,annot=True,cmap="CMRmap_r",fmt="d")

plt.subplot(2,3,5)
plt.title("Random Forest Confusion Matrix")
sns.heatmap(cm_rfc,cbar=False,annot=True,cmap="CMRmap_r",fmt="d")

# plt.subplot(2,3,6)
# plt.title("Support Vector Machine Confusion Matrix")
# sns.heatmap(cm_svm,cbar=False,annot=True,cmap="CMRmap_r",fmt="d")

plt.subplot(2,3,6)
plt.title("Support Vector Machine Linear Confusion Matrix")
sns.heatmap(cm_svm_linear,cbar=False,annot=True,cmap="CMRmap_r",fmt="d")

plt.show()
```

```python
algorithms = ("Logistic Regression","Decision Tree","Random Forest","K Nearest Neighbors","Naive Bayes","SVM-Kernel:Poly", "SVM-Kernel:Linear")
scores = (lr_score,dc_score,rfc_score,knn_score,nb_score,svm_score,svm_score_linear)
y_pos = np.arange(1,8)
colors = ("red","gray","purple","green","orange","blue","yellow")

plt.figure(figsize=(24,12))
plt.xticks(y_pos,algorithms,fontsize=18)
plt.yticks(np.arange(0.00, 1.01, step=0.01))
plt.ylim(0.90,1.00)
plt.bar(y_pos,scores,color=colors)
plt.grid()
plt.suptitle("Acc dos modelos",fontsize=24)
plt.show()

```
